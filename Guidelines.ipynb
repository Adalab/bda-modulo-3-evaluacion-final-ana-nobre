## 📁 Los Datos

Los datos que se proporcionan consisten en dos conjuntos de archivos que, en conjunto, describen el comportamiento de los clientes dentro de un programa de lealtad de una aerolínea.

### ✈️ `Customer Flight Activity.csv`

Contiene información sobre la actividad de vuelo de los clientes, incluyendo vuelos reservados, distancia volada, puntos acumulados y redimidos, y costos asociados:

- **Loyalty Number**: Identificador único para cada cliente.
- **Year**: Año del registro.
- **Month**: Mes (1 a 12).
- **Flights Booked**: Número de vuelos reservados ese mes.
- **Flights with Companions**: Vuelos con acompañantes.
- **Total Flights**: Total acumulado de vuelos del cliente.
- **Distance**: Distancia total volada ese mes.
- **Points Accumulated**: Puntos ganados durante el mes.
- **Points Redeemed**: Puntos redimidos durante el mes.
- **Dollar Cost Points Redeemed**: Valor en dólares de los puntos redimidos.

### 👤 `Customer Loyalty History.csv`

Proporciona un perfil detallado de los clientes:

- **Loyalty Number**: Identificador único para correlacionar con el archivo de actividad.
- **Country**, **Province**, **City**, **Postal Code**
- **Gender**, **Education**, **Salary**, **Marital Status**
- **Loyalty Card**: Tipo de tarjeta de lealtad.
- **CLV**: Customer Lifetime Value.
- **Enrollment Type**, **Enrollment Year**, **Enrollment Month**
- **Cancellation Year**, **Cancellation Month** (si aplica)

## 🧪 Ejercicio

### 🧼 Fase 1: Exploración y Limpieza

1. **Exploración Inicial**:  
   - Detectar valores nulos, atípicos o inconsistentes.  
   - Usar pandas para conocer la estructura y estadísticas básicas.  
   - Unir los datasets de forma eficiente.

2. **Limpieza de Datos**:  
   - Eliminar o tratar valores nulos en columnas clave.  
   - Verificar consistencia y corregir errores.  
   - Convertir tipos de datos según necesidad.

---

### 📊 Fase 2: Visualización

Usando las herramientas vistas en el módulo, responder gráficamente:

1. ¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año?  
2. ¿Existe una relación entre la distancia volada y los puntos acumulados?  
3. ¿Cuál es la distribución de clientes por provincia o estado?  
4. ¿Cómo se compara el salario promedio entre niveles educativos?  
5. ¿Cuál es la proporción de tipos de tarjetas de lealtad?  
6. ¿Cómo se distribuyen los clientes según estado civil y género?

---

### 🎯 BONUS - Fase 3: Diferencias por Nivel Educativo

Evaluar si hay diferencias significativas en el número de vuelos reservados según el nivel educativo:

1. **Preparación de datos**: usar solo columnas `Flights Booked` y `Education`.  
2. **Análisis Descriptivo**: promedio, desviación estándar por grupo.  
3. **Prueba Estadística**: aplicar test de hipótesis para detectar diferencias.
